```python
import json
from matplotlib import pyplot as plt

def count_chickens(json_file_path):
    with open(json_file_path, 'r') as file:
        data = json.load(file)

    return data
json_file_path = "human_keypoints_with_activity.json"
data =  count_chickens(json_file_path)

keypoint_index1 = []
activity = []
act_label = ["sitting", "standing", "sleeping"]

print(type(data))
for img_idx in range(0, len(data)):
    keypoint_index1.append(data[img_idx]["keypoints"][0])
    if data[img_idx]["activity"]== act_label[0]:
        activity.append(1)
    elif data[img_idx]["activity"]== act_label[1]:
        activity.append(2)
    elif data[img_idx]["activity"]== act_label[2]:
        activity.append(3)
    

print(keypoint_index1)


# Use scatter plot
plt.scatter(keypoint_index1, activity, c=activity, cmap="viridis", alpha=0.7)
plt.xlabel('Keypoint Index 1')
plt.ylabel('Activity')

# Replace numeric y-ticks with activity labels
plt.yticks([1, 2, 3], act_label)

plt.title('Keypoint Index 1 vs Activity')
plt.show()
```
